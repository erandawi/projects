```python
def data_summary(df):
    summary = pd.DataFrame({
        'Data Type': df.dtypes,
        'Missing Values': df.isnull().sum(),
        'Missing %': df.isnull().mean() * 100,
        'Unique Values': df.nunique(),
        'Sample Values': df.apply(lambda x: x.unique()[:5] if x.nunique() < 20 else x.sample(5).unique())
    })

    summary['Missing %'] = summary['Missing %'].round(2)
    summary = summary.sort_values(by='Missing %', ascending=False)
    
    return summary
```

```python
data_summary(df)
```

### Univariate Analysis

```python
df[df.columns].nunique()
```

```python
# split columns based on feature types
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include='object').columns
print(f"numerical columns: {num_cols}")
print(f"categorical columns: {cat_cols}")
```

```python
# Continuos Features
# histogram + KDE
for col in num_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df[col].dropna(),kde = True, bins = 30)
    plt.title(f"{col} Distribution")
    plt.show()
```

```python
# Boxplot for Outliers
for col in num_cols:
    plt.figure(figsize=(6, 1.5))
    sns.boxplot(x=df[col])
    plt.title(f"{col} â€“ Outliers Check")
    plt.show()
```

```python
# Skewness Check
df[num_cols].skew().sort_values(ascending=False)
```

```python
# Categorical Features
# Value Counts + % Contribution
for col in cat_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts(dropna=False))
    print("Unique values:", df[col].nunique())
    print("Missing values:", df[col].isna().sum())
```

```python
# Bar Plots for Categorical
for col in cat_cols:
    if df[col].nunique() < 30:  # Skip high cardinality
        plt.figure(figsize=(8, 4))
        df[col].value_counts().plot(kind='bar')
        plt.title(f"{col} â€“ Frequency")
        plt.xticks(rotation=45)
        plt.show()
```
