```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

```python
dataset = pd.read_csv("data/Algerian_forest_fires_dataset.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

## Data Cleaning

```python
# Missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets based on Region from 12th index, we can make a new column based on the Region

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes Region Dataset"

Add new column with region

```python
dataset.loc[:122, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# Remove the 122nd row
df = df.drop(index=[122]).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# fix spaces in columns names
df.columns = df.columns.str.strip()
df.columns
```

```python
df['Classes'] = df['Classes'].str.strip()
```

```python
df.info()
```

## Change the required columns as integer data type

```python
df.columns
```

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
objects = [features for features in df.columns if df[features].dtype == 'object']
```

```python
objects
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype('float')
```

```python
df.info()
```

```python
objects
```

```python
df.describe()
```

```python
df.head()
```

```python
# Saved clean dataset
```

```python
df.to_csv('data/Algerian_forest_fires_dataset_cleaned.csv', index =False)
```

## Exploratory Data Analysis

```python
## drop day month and year
df_copy = df.drop(['day', 'month', 'year'], axis=1)
```

```python
df_copy.head()
```

```python
# categories in classes
df_copy['Classes'].value_counts()
```

```python
## Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy.tail()
```

```python
df_copy['Classes'].value_counts()
```

```python
## plot density of all features
plt.style.use('ggplot')
df_copy.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Percentage for pie chart
percentage = df_copy['Classes'].value_counts(normalize=True) * 100
```

```python
# plotting piechart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels = classlabels, autopct = '%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
